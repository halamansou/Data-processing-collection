```python
import numpy as np

import pandas as pd

data = pd.read_csv("E:\semester7\diabetics2.csv")
data
```

```python
##Missing Values
from sklearn.impute import SimpleImputer

impute=SimpleImputer(missing_values=np.nan , strategy="mean")

impute.fit(data.iloc[: , 1:3])

data.iloc[: , 1:3]=impute.transform(data.iloc[: , 1:3])
data
```

```python
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer([(" ", OneHotEncoder(),[0])],remainder='passthrough')

data=pd.DataFrame(ct.fit_transform(data))
data
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
data.iloc[: , -1] =le.fit_transform(data.iloc[: ,-1])
data

```
