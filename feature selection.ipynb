```python
import numpy as np

import pandas as pd

data = pd.read_csv("E:\semester7\diabetics.csv")

print (data.shape)

data
```

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x=pd.DataFrame(st.fit_transform(x))
print(x)
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
x=pd.DataFrame(mm.fit_transform(x))
x
```

```python
from sklearn.feature_selection import RFE
from sklearn import svm
model=svm.SVC(kernel='linear')
rfe=RFE(model,n_features_to_select=10)
rfe.fit(x, y)
filter = rfe.support_
ranking = rfe.ranking_
print(filter)
print("-------------------------------------------------")
print(ranking)
print("-------------------------------------------------")

data=x[x.columns[filter]]
data["outcome"]=y
data.to_csv("newdata.csv",index=False)
print("Data shape = " , data.shape)
print("-------------------------------------------------")
data
```

```python
data = pd.read_csv("newdata.csv")
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
model=svm.SVC(kernel='rbf')
#X_train, X_test, y_train, y_test = train_test_split( test_size=0.2, stratify=X_test , random_state=42)
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2 , random_state=42)
model.fit(X_train, y_train)
predections=model.predict(X_test)
from sklearn.metrics import accuracy_score
print(data.shape)
print(accuracy_score(y_test, predections))
```
